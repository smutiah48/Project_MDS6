```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
```

```python

# Ganti dengan API key Anda.
api_key = 'AIzaSyDStK-DXk13eTOb6s1DJzsF4ZRoh-YC_GU'

# Contoh pencarian buku dengan judul "Statistik".
search_term = "Pengembangan diri"
max_results = 40  # Maksimum adalah 40
url = f"https://www.googleapis.com/books/v1/volumes?q={search_term}&maxResults={max_results}&key={api_key}"

# Melakukan permintaan ke Google Books API.
response = requests.get(url)
data = response.json()

# Siapkan list untuk menampung data buku.
books = []

# Analisis data dan ekstraksi informasi jika item tersedia.
if 'items' in data:
    for item in data['items']:
        book_id = item['id']
        book_title = item['volumeInfo']['title']
        book_authors = item['volumeInfo'].get('authors', [])
        book_categories = ', '.join(item['volumeInfo'].get('categories', [])) if 'categories' in item['volumeInfo'] else 'Kategori tidak tersedia'
        book_publisher = item['volumeInfo'].get('publisher', 'Penerbit tidak tersedia')
        book_published_date = item['volumeInfo'].get('publishedDate', 'Tanggal terbit tidak tersedia')
        book_page_count = item['volumeInfo'].get('pageCount', 'Jumlah halaman tidak tersedia')
        book_isbn = [isbn['identifier'] for isbn in item['volumeInfo'].get('industryIdentifiers', []) if isbn['type'] == 'ISBN_13']
        book_isbn = book_isbn[0] if book_isbn else 'ISBN tidak tersedia'
        book_link = item['volumeInfo'].get('infoLink', 'Link buku tidak tersedia')
        book_description = item['volumeInfo'].get('description', 'Deskripsi tidak tersedia')
        # Tempat penerbit sebagai placeholder
        book_publisher_location = "Informasi tidak tersedia"

        # Menambahkan data ke list books.
        books.append({
            'ID Buku': book_id,
            'Judul': book_title,
            'Penulis': ', '.join(book_authors),
            'Kategori': book_categories,
            'Nama Penerbit': book_publisher,
            'Tempat Penerbit': book_publisher_location,  # Placeholder untuk lokasi penerbit
            'Tahun Terbit': book_published_date,
            'Jumlah Halaman': book_page_count,
            'ISBN': book_isbn,
            'Link Buku': book_link,
            'Deskripsi': book_description
        })

    # Membuat DataFrame dari list buku.
    books_df = pd.DataFrame(books)

    # Menampilkan DataFrame.
    display(books_df)
else:
    print("Tidak ada hasil yang ditemukan.")

# Jika ingin menyimpan ke file CSV:
# books_df.to_csv('books.csv', index=False)
```
